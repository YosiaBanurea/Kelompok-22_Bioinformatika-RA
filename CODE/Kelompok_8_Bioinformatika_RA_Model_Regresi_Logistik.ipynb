```python
# Drop rows where 'potential_duplicate' column has a value of 1
data = data[data['potential_duplicate'] != 1]
data
```

## UNDERSAMPLING UNTUK BALANCED DATA TARGET

```python
# Check the distribution of values in the 'assay_type' column
assay_type_distribution = data['assay_type'].value_counts()

# Perform undersampling to match the count of 'B' to 'F'
min_count = assay_type_distribution.min()

# Separate the dataset by 'assay_type'
df_B = data[data['assay_type'] == 'B']
df_F = data[data['assay_type'] == 'F']

# Perform undersampling
df_B_undersampled = df_B.sample(n=min_count, random_state=42)
df_F_undersampled = df_F.sample(n=min_count, random_state=42)

# Combine the undersampled datasets
df_balanced = pd.concat([df_B_undersampled, df_F_undersampled])

# Shuffle the combined dataset
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# Save the balanced dataset to a new file
Target_Balanced = 'Target_Balanced.csv'
df_balanced.to_csv(Target_Balanced, index=False)

# Provide the path to the balanced file for download
Target_Balanced
```

## Buat fingerprint dari canonical_smiles

```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import AllChem
from rdkit.Chem import rdMolDescriptors
```

```python
#  Buat fingerprint dari canonical_smiles (Still using AllChem for Morgan Fingerprints)
def generate_fingerprint(smiles, radius=2, nBits=2048):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return None
    # Generate Morgan Fingerprint using AllChem
    fingerprint = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=nBits)
    return fingerprint

# Suppress deprecation warnings temporarily (optional)
#warnings.filterwarnings("ignore", category=DeprecationWarning)

Target_Balanced['fingerprint'] = Target_Balanced['canonical_smiles'].apply(generate_fingerprint)

# Hapus baris dengan fingerprint None
df_resampled = Target_Balanced.dropna(subset=['fingerprint'])

# Ubah fingerprint menjadi array numpy
X = [list(fp) for fp in df_resampled['fingerprint']]

# Tentukan variabel target
y = df_resampled['pchembl_value']

df_resampled
```

## ubah fingerprint ke format one hot encoding

```python
import pandas as pd
from rdkit import DataStructs
import numpy as np

def fingerprint_to_onehot(fp):
    arr = np.zeros((1,))
    DataStructs.ConvertToNumpyArray(fp, arr)
    return arr

# Apply the conversion to the 'fingerprint' column
df_resampled['fingerprint_onehot'] = df_resampled['fingerprint'].apply(fingerprint_to_onehot)

# Expand one-hot encoded arrays into separate columns
onehot_df = pd.DataFrame(df_resampled['fingerprint_onehot'].tolist())

# Concatenate with the original dataframe
df_final = pd.concat([df_resampled, onehot_df], axis=1)

# Save to a CSV file
df_final.to_csv('df_with_onehot.csv', index=False)
```

## isi nilai pchembl_value dengan nilai kosong

```python
# Fill NaN values in the 'pchembl_value' column with the column's mean
df_final['pchembl_value'] = df_final['pchembl_value'].fillna(df_final['pchembl_value'].mean())
df_final
```

## klasifikasikan nilai activity_comment berdasarkan pchembl_value

```python
# Classify 'activity_comment' based on 'pchembl_value'
df_final['activity_comment'] = df_final['pchembl_value'].apply(
    lambda x: 'active' if x > 5 else 'inactive'
)

df_final
```

```python
df_final['activity_comment'].value_counts()
```

## Ubah Nilai pada kolom assay_type menjadi 0 dan 1

```python
# Assign 1 to 'B' and 0 to 'F'
df_final['assay_type_binary'] = df_final['assay_type'].apply(lambda x: 1 if x == 'B' else 0)
df_final
```

```python
df_final['assay_type_binary'].value_counts()
```

## ubah nilai pada kolom activity_comment menjadi 0 dan 1

```python
# Replace 'active' with 1 and 'inactive' with 0 in the 'activity_comment' column
df_final['activity_comment'] = df_final['activity_comment'].replace({'active': 1, 'inactive': 0})
df_final
```

```python
df_final.to_csv('df_final.csv', index=False)
```

## MODELING DENGAN MODEL REGRESI LOGISTIK

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
import numpy as np
```

```python
features = df_final.drop(columns=['molecule_chembl_id', 'canonical_smiles', 'bao_format', 'fingerprint_onehot', 'assay_type','assay_type_binary', 'fingerprint'])
target = df_final['assay_type_binary']
```

```python
# Convert all column names to strings
features.columns = features.columns.astype(str)

# Now you can proceed with splitting and training:
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Train the logistic regression model
logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = logistic_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Generate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=logistic_model.classes_)
disp.plot(cmap=plt.cm.Blues)

plt.title("Confusion Matrix")
plt.show()

# Generate classification report
report = classification_report(y_test, y_pred)
print(report)
```
